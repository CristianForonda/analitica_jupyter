```python
# METODOS COEFICIENTE ORGANIZACIONAL ------------------

# Calcular moda para Coeficiente Analitica
def modaCOrganizacional(df):

    x1 = df["C25"]
    x2 = df["C26"]
    x3 = df["C28"]
    x4 = df["C31"]
    x5 = df["C32a"]

    df2 = [x1, x2, x3, x4, x5]
    # definir un diccionario para el conteo de numeros
    dconteo = {}
    for j in df2:
        key = str(j)
        # Si este no existe
        if not key in dconteo:
            # se agrega al diccionario
            dconteo[key] = 1
        # si existe en el diccionario se suma
        else:
            # se cuenta
            dconteo[key] += 1

    # Se recorre el diccionario y se obtiene el mayor frecuencia
    mayor_frec = 0
    num_mas_repetido = df2[0]

    for j in dconteo:
        if dconteo[j] > mayor_frec:
            num_mas_repetido = j
            mayor_frec = dconteo[j]

    conteo = dconteo[str(num_mas_repetido)]

    # le hago un cast a entero, debido a que lo devuelve como un caracter
    return int(num_mas_repetido)
```

```python
# Calcular laa pregunta c25,c26,c28,c31 y definir la c32  en blanco para luego agregarla
def calcularCalificacionOrganizacional(df):
    # inicializar variables en 0 para que no guarden basura de pruebas anteriores
    cal_c25 = 0
    cal_c26 = 0
    cal_c28 = 0
    cal_c31 = 0
    cal_c32 = 0
    cal32a = "NaN"
    listapOrga = []
    # metodo para calcular las calificaciones,  preguntas con su token , respuesta y calificacion asignada
    for i in range(len(df)):

        # PREGUNTA 25
        if df[qr.r25][i] == qr.r25a:
            cal_c25 = 1

        elif df[qr.r25][i] == qr.r25b:
            cal_c25 = 2

        elif df[qr.r25][i] == qr.r25c:
            cal_c25 = 3

        elif df[qr.r25][i] == qr.r25d:
            cal_c25 = 4

        elif df[qr.r25][i] == qr.r25e:
            cal_c25 = 5

        else:
            cal_c25 = 0

        # PREGUNTA 26
        if df[qr.r26][i] == qr.r26a:
            cal_c26 = 1

        elif df[qr.r26][i] == qr.r26b:
            cal_c26 = 2

        elif df[qr.r26][i] == qr.r26c:
            cal_c26 = 3

        elif df[qr.r26][i] == qr.r26d:
            cal_c26 = 4

        elif df[qr.r26][i] == qr.r26e:
            cal_c26 = 5

        else:
            cal_c26 = 0

        # PREGUNTA 28
        if df[qr.r28][i] == qr.r28a:
            cal_c28 = 1

        elif df[qr.r28][i] == qr.r28b:
            cal_c28 = 2

        elif df[qr.r28][i] == qr.r28c:
            cal_c28 = 3

        elif df[qr.r28][i] == qr.r28d:
            cal_c28 = 4

        elif df[qr.r28][i] == qr.r28e:
            cal_c28 = 5

        else:
            cal_c28 = 0

        # PREGUNTA 31
        if df[qr.r31][i] == qr.r31a:
            cal_c31 = 1

        elif df[qr.r31][i] == qr.r31b:
            cal_c31 = 2

        elif df[qr.r31][i] == qr.r31c:
            cal_c31 = 3

        elif df[qr.r31][i] == qr.r31d:
            cal_c31 = 4

        elif df[qr.r31][i] == qr.r31e:
            cal_c31 = 5

        else:
            cal_c31 = 0

        if pd.isnull(df[qr.r32a][i]):
            cal32a = 0
        else:
            cal32a = df[qr.r32a][i]

        # Columna B   [Nivel operativo/ táctico]
        if pd.isnull(df[qr.r32b][i]):
            cal32b = 0
        else:
            cal32b = df[qr.r32b][i]

        # Columna C [Áreas de tecnologías]
        if pd.isnull(df[qr.r32c][i]):
            cal32c = 0
        else:
            cal32c = df[qr.r32c][i]

        # la pregunta c32 es regla (?) tiene 3 opciones, para autoevaluarse

        # genero d1 con las calificaciones
        d1 = dict(
            [
                ("Token", df.loc[i, "Token"]),
                ("C25", cal_c25),
                ("C26", cal_c26),
                ("C28", cal_c28),
                ("C31", cal_c31),
                ("C32a", int(cal32a)),
                ("C32b", int(cal32b)),
                ("C32c", int(cal32c)),
            ]
        )

        # asigno la moda a un nuevo dataframe
        d25 = dict(
            [
                ("Token", df.loc[i, "Token"]),
                ("C25", cal_c25),
                ("C26", cal_c26),
                ("C28", cal_c28),
                ("C31", cal_c31),
                ("C32a", int(cal32a)),
                ("C32b", int(cal32b)),
                ("C32c", int(cal32c)),
                ("MODA_D", modaCOrganizacional(d1)),
            ]
        )
        print(i, "d25", d25)

        listapOrga.append(d25)

    return listapOrga
```
