```python
# Calcular moda
def moda(df):

    x1 = df["C1"]
    x2 = df["C2"]
    x3 = df["C3"]
    x4 = df["C4"]
    x5 = df["C5"]
    x8 = df["C8"]

    df2 = [x1, x2, x3, x4, x5, x8]

    # definir un diccionario para el conteo de numeros
    dconteo = {}
    for j in df2:
        key = str(j)
        # Si este no existe
        if not key in dconteo:
            # se agrega al diccionario
            dconteo[key] = 1
        # si existe en el diccionario se suma
        else:
            # se cuenta
            dconteo[key] += 1

    # Se recorre el diccionario y se obtiene el mayor frecuencia
    mayor_frec = 0
    num_mas_repetido = df2[0]
    # se imprime el diccionario  con la cantidad de repeticiones
    # Y sacamos el mayor
    for j in dconteo:
        if dconteo[j] > mayor_frec:
            num_mas_repetido = j
            mayor_frec = dconteo[j]

    conteo = dconteo[str(num_mas_repetido)]
    # le hago un cast a entero, debido a que lo devuelve como un caracter
    return int(num_mas_repetido)
```
