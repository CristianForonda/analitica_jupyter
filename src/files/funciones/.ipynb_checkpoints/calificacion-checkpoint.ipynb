```python
# calculo de calificaciones segun la rubrica pregunta 5
def calcularCalp5(cantsi):
    calificacion = 0
    if cantsi == 4:
        calificacion = 5
    elif cantsi == 3:
        calificacion = 4
    elif cantsi == 2:
        calificacion = 3
    elif cantsi == 1:
        calificacion = 2
    else:
        calificacion = 0

    return calificacion
```

```python
# Calcular calificaci칩n pregunta 5 Escala
def calcularCalificacionEscala(df):
    # metodo para calcular las calificaciones, preguntas con su token , respuesta y calificacion asignada a la pregunta
    listap5 = []
    for i in range(len(df)):
        cantsi = 0
        # primeraPregunta
        if df[qr.r5a][i] == "Si":
            cantsi += 1

        if df[qr.r5b][i] == "Si":
            cantsi += 1

        if df[qr.r5c][i] == "Si":
            cantsi += 1

        if df[qr.r5d][i] == "Si":
            cantsi += 1

        dp5 = dict([("Token", df.loc[i, "Token"]), ("C5", calcularCalp5(cantsi))])

        listap5.append(dp5)

    return listap5
```
