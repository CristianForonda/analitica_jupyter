```python
# Calcular laa pregunta 17 y definir las otras en blanco para luego agregarlas
def calcularCalificacionC17(df):
    # inicializar variables en 0 para que no guarden basura de pruebas anteriores
    cal_c17 = 0
    cal_c19 = 0
    cal_c20 = 0
    cal_c21 = 0
    cal_c22 = 0
    cal_c24 = 0
    listap17 = []
    # metodo para calcular las calificaciones, primera pregunta con su token , respuesta y calificacion asignada a la p1
    for i in range(len(df)):

        # PREGUNTA 17
        if df[qr.r17][i] == qr.r17a:
            cal_c17 = 1

        elif df[qr.r17][i] == qr.r17b:
            cal_c17 = 2

        elif df[qr.r17][i] == qr.r17c:
            cal_c17 = 3

        elif df[qr.r17][i] == qr.r17d:
            cal_c17 = 4

        elif df[qr.r17][i] == qr.r17e:
            cal_c17 = 5

        else:
            cal_c17 = 0

        # las otras preguntas son escalas
        # genero d17 con las calificaciones
        d1 = dict(
            [
                ("Token", df.loc[i, "Token"]),
                ("C17", cal_c17),
                ("C19", cal_c19),
                ("C20", cal_c20),
                ("C21", cal_c21),
                ("C22", cal_c22),
                ("C24", cal_c24),
            ]
        )

        # asigno la moda a un nuevo dataframe
        d17 = dict(
            [
                ("Token", df.loc[i, "Token"]),
                ("C17", cal_c17),
                ("C19", cal_c19),
                ("C20", cal_c20),
                ("C21", cal_c21),
                ("C22", cal_c22),
                ("C24", cal_c24),
                ("MODA_C", calculo.modaCAnalitica(d1)),
            ]
        )
        # print(i, "d2", d17)

        listap17.append(d17)

    return listap17
```

```python
# Calcular calificación pregunta Escala c19
def calcularCalificacionEscalaC19(df):
    # metodo para calcular las calificaciones, preguntas con su token , respuesta y calificacion asignada a la pregunta
    listac19 = []
    for i in range(len(df)):
        cal19 = 0
        # se podria poner un elif para No y para N/A pero igual no suman calificación
        # Descriptiva
        if df[qr.r19a][i] == "Yes":
            cal19 = 2

        # Diagnostica
        if df[qr.r19b][i] == "Yes":
            cal19 = 3

        # Predictiva
        if df[qr.r19c][i] == "Yes":
            cal19 = 4

        # Prescriptiva
        if df[qr.r19d][i] == "Yes":
            cal19 = 5

        # Ninguna de las mencionadas
        if df[qr.r19e][i] == "Yes":
            cal19 = 1

        dp19 = dict([("Token", df.loc[i, "Token"]), ("C19", cal19)])

        listac19.append(dp19)

    return listac19
```
